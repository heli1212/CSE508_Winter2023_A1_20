```python
import os
from nltk.tokenize import word_tokenize
import pickle
```

```python
# Folder Path
path = "D:\SEMESTER-2\IR-Ass1\Dataset"

# Change the directory
os.chdir(path)
```

```python
dict = {}
dict2 = {}
count=[0]
def tokenize_file(file_path):   
    with open(file_path, 'r') as f:
        for line in f:
            count[0]=count[0]+1
            temp=word_tokenize(line)

            for item in temp:
                if item not in dict:
                    dict[item] = []
                    dict2[item] = 0
                if item in dict:
                    dict[item].append(count[0])
                    temp_arr=sorted([*set(dict[item])])
                    dict[item]=temp_arr
                    dict2[item] = dict2[item] + 1

for file in os.listdir():  
    file_path = f"{path}\{file}"

    # call tokenize file function
    tokenize_file(file_path)

for i in dict:
    print("{x}: {dict2} -> {dict1}".format(x=i,dict2=dict2[i],dict1=dict[i]))
print(len(dict2))
```

```python
## AND Operation

str1 = 'slip'
str2 = 'appropriate'

list1 = dict[str1]
list2 = dict[str2]

print(list(set(list1) & set(list2)))
```

```python
## OR Operation

str1 = 'slip'
str2 = 'appropriate'

list1 = dict[str1]
list2 = dict[str2]

print(list(set(list1) | set(list2)))
# print(list1)
# print(list2)
```

```python
## AND NOT Operation
str1 = 'experiment'
str2 = 'aerodynamics'

list1 = dict[str1]
list2 = dict[str2]

notList = []
for i in range(1,1401):
    if i not in list2:
        notList.append(i)

print(list(set(list1) & set(notList)))
```
