```python
def readfile(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            print(line)
```

```python
# Read text File
def read_text_file(file_path):
    with open(file_path, 'r') as f:
        s=""
        for line in f:
            s+=line.strip()
            
        tag = "TITLE"
        tag2="TEXT"

        reg_str = "<" + tag + ">(.*?)</" + tag + ">"
        reg_str2 = "<" + tag2 + ">(.*?)</" + tag2 + ">"

        res = re.findall(reg_str,s)
        res2 = re.findall(reg_str2,s)

    with open (file_path, 'w') as file:  
        file.write(str(res[0])+str(res2[0]))  

## Print only 5 files 
c=6
for file in os.listdir():
    c=c-1
    if c==0:
        break
    file_path = f"{path}\{file}"
    print("FILE: "+file)
    readfile(file_path)
    print("")
```

```python
# Lower Case
def lowercase_file(file_path):
    with open(file_path, 'r') as f:
        s=""
        for line in f:
            s+=line.strip()    
        s.lower()

    with open (file_path, 'w') as file:  
        file.write(s)  
```

```python
#  TOKENIZATION  #

def tokenize_file(file_path):   
    with open(file_path, 'r') as f:
        for line in f:
            temp=word_tokenize(line)
    
```

```python
## Remove Stop Words

def removeStopwords(file_path):
    stoWords=set(stopwords.words('english'))
    s=""
    with open(file_path, 'r') as f:
        ans=[]
        
        for line in f:
            temp=word_tokenize(line)
            for h in temp:
                if not h in stoWords:
                    ans.append(h)
            for i in ans:
                s+=i+" "

    with open(file_path, 'w') as f:
        f.write(s)
```

```python
## Remove Puntuations

def removePuntuations(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            test_str = line.translate(str.maketrans('', '', string.punctuation))

    with open(file_path, 'w') as f:
        f.write(test_str)
```

```python
### Remove White spaces

def removeWS(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            s1=" ".join(line.split())
            
    with open(file_path, 'w') as f:
        f.write(s1)
```

```python
### Function Calls
        
for file in os.listdir():  
    file_path = f"{path}\{file}"
    
    lowercase_file(file_path)
    tokenize_file(file_path)
    removeStopwords(file_path)
    removePuntuations(file_path)
    removeWS(file_path)
```

```python
# Prints the file        

def print_file():
    c=6
    for file in os.listdir():
        c=c-1
        if c==0:
            break
        file_path = f"{path}\{file}"
        print("FILE: "+file )
        readfile(file_path)
        print("")
        
print_file()
```
